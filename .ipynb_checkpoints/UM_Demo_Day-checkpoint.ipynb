Column information:

Err column 1 is positive error, Err column 2 is negative error

koi_disposition: Disposition Using Kepler Data. The disposition in the literature towards this exoplanet candidate. 
One of CANDIDATE, FALSE POSITIVE, NOT DISPOSITIONED or CONFIRMED.

koi_pdisposition: The disposition Kepler data analysis has towards this exoplanet candidate. 
One of FALSE POSITIVE, NOT DISPOSITIONED, and CANDIDATE.

koi_score: A value between 0 and 1 that indicates the confidence in the KOI disposition. For CANDIDATEs, a higher value indicates more confidence in its disposition, while for FALSE POSITIVEs, a higher value indicates less confidence in that disposition. The value is calculated from a Monte Carlo technique such that the score's value is equivalent to the frction of iterations where the Robovetter yields a disposition of CANDIDATE.

koi_fpflag_nt: Not Transit-Like Flag. The light curve is not consistent with that of a transiting planet.

koi_fpflag_ss: Stellar Eclipse Flag. A KOI that is observed to have a significant secondary event, transit shape, 
or out-of-eclipse variability, which indicates that the transit-like event is most likely caused by an eclipsing binary.

koi_fpflag_co: Centroid Offset Flag. The source of the signal is from a nearby star, as inferred by measuring the 
centroid location of the image both in and out of transit.

koi_fpflag_ec: Ephemeris Match Indicates Contamination Flag. The KOI shares the same period and epoch as another object 
and is judged to be the result of flux contamination in the aperture or electronic crosstalk.

koi_period: Orbital Period (days). The interval between consecutive planetary transits.

koi_time0bk: Transit Epoch (BJD - 2,454,833.0). The time corresponding to the center of the first detected transit in 
Barycentric Julian Day (BJD) minus a constant offset of 2,454,833.0 days. 
The offset corresponds to 12:00 on Jan 1, 2009 UTC.

koi_impact: Impact Parameter. The sky-projected distance between the center of the stellar disc and the center of the 
planet disc at conjunction, normalized by the stellar radius.

koi_duration: Transit Duration (hours). The duration of the observed transits.

koi_depth: Transit Depth (parts per million).The fraction of stellar flux lost at the minimum of the planetary transit.

koi_prad: Planetary Radius (Earth radii). The radius of the planet. Planetary radius is the product of the planet star 
radius ratio and the stellar radius.

koi_teq: Equilibrium Temperature (Kelvin). Approximation for the temperature of the planet.

koi_insol: Insolation Flux [Earth flux]. Insolation flux is another way to give the equilibrium temperature.

koi_model_snr: Transit Signal-to-Noise. Transit depth normalized by the mean uncertainty in the flux during the transits.

koi_tce_plnt_num: TCE Planet Number. TCE Planet Number federated to the KOI.

koi_steff: Stellar Effective Temperature (Kelvin). The photospheric temperature of the star.

koi_slogg: Stellar Surface Gravity (log10(cm s-2). The base-10 logarithm of the acceleration due to gravity at the 
surface of the star.

koi_srad: Stellar Radius (solar radii). The photospheric radius of the star.

ra & dec: right ascension and declination (degrees).

koi_kepmag: Kepler-band (mag). Kepler-band (mag).


```python
# Drop columns like rowid, kepler_name, and the RA and Dec
exoplanet_df = exoplanet_df.drop(['rowid', 'koi_pdisposition','kepid', 'kepoi_name', 'kepler_name', 
                                  'koi_tce_delivname', 'ra', 'dec'], axis=1)

exoplanet_df.head()
```

```python
# Check to see which columns have and NaN
exoplanet_df.isna().any()
```

```python
# Most do. Dropping the NaN values

# Drop the null columns where all values are null
exoplanet_df = exoplanet_df.dropna(axis='columns', how='all')

# Drop the null rows
exoplanet_df = exoplanet_df.dropna()

exoplanet_df.head()
```

```python
# See details of cleaned data set
exoplanet_df.describe()
```

## LabelEncode the categorical data

```python
# koi_disposition is our label
y = exoplanet_df.koi_disposition
```

```python
# Need to convert strings to integer values
from sklearn.preprocessing import LabelEncoder

# Label-encode the data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)

encoded_y
```

```python
# Label encoding turns Candidate exoplanets into 0, Confirmed into 1, and False Positives into 2

for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
# Add the encoded_y to the dataframe
exoplanet_df['encoded_y'] = encoded_y

# Drop the koi_disposition
exoplanet_df = exoplanet_df.drop(['koi_disposition'], axis=1)
```

# Selecting the features

I know I'm going to need more than just the four flag columns. They give about upper-60 to lower-70 percent results, with a bit too high of loss for my own taste. I plan on using temperature, and since that can vary widely, and is much higher than the flag values, I will normalize the data.

```python
# Normalize the data using z-scores to get the values to be in about the same range
# Calculate the Z-scores of each column in the training set:

exoplanet_df_mean = exoplanet_df.mean()
exoplanet_df_std = exoplanet_df.std()
exoplanet_df_norm = (exoplanet_df - exoplanet_df_mean)/exoplanet_df_std

exoplanet_df_norm
```

After testing, I find the following features perform best: in addition to the flags, the depth of transit, the stellar temperature, and the temperature of the possible planet will be what I use. These make sense to me: the majority of stars are binary, so higher temperatures for the object mean more a chance at the object being an eclipsing binary. The stellar temperature is directly related to the star's luminosity, so a higher temperature means it's easier to see an eclipsing binary occur. The depth deals with stellar flux, which is again tied to luminosity and ease in spotting an object transiting the star.

```python
# Set features

selected_features = exoplanet_df_norm[['koi_fpflag_nt','koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec',
                                      'koi_depth', 'koi_steff', 'koi_teq']]
```

# Split the data into training and testing sets

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

# random_state shuffles the data
X_train, X_test, y_train, y_test = train_test_split(selected_features, encoded_y, random_state=1)

X_train.head()
```

# Data Pre-processing

```python
# Scale the data using the StandardScaler
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## One-hot Encoding

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_train_categorical
```

# Train the Model



```python
# Want to visualize the loss to make it easier to check
def plot_the_loss_curve(epochs, loss, accuracy):
    """Plot a curve of loss vs. epoch."""

    plt.figure()
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy/Loss")

    plt.plot(epochs, accuracy, label="Accuracy")
    plt.plot(epochs, loss, label="Loss")
    plt.legend()
    plt.show()
```

## Defining our Model Architecture (the layers)


```python
tf.keras.backend.clear_session()
```

```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = len(selected_features.columns)
number_hidden_nodes = 5
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
number_classes = 3
model.add(Dense(units=number_classes, activation='softmax', kernel_regularizer='l2'))
```

```python
model.summary()
```

```python
my_learning_rate = 0.009
# Construct the layers into a model that TensorFlow can execute.
model.compile(optimizer= tf.keras.optimizers.Adam(lr=my_learning_rate),
              loss='MeanSquaredError',
              metrics=['accuracy'])
```

```python
# Fit (train) the model
history = model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    batch_size=35,
    shuffle=True,
    verbose=1
)

# Get details that will be useful for plotting the loss curve.
epochs = history.epoch
hist = pd.DataFrame(history.history)
accuracy = hist.accuracy
loss = hist.loss

# Plot the accuracy over time
plot_the_loss_curve(epochs, loss, accuracy)
```

```python
# Evaluate the model using the testing data
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

I have reduced my loss, while at the same time increasing my accuracy to nearly (sometimes above) 80 percent. With the time I have, these are okay results.
